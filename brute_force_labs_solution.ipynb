## Naive implementation of LABS problem to generate simple 2D plot representing function


```python
from typing import Tuple

import matplotlib.pyplot as plt
```

```python
def to_binary_tuple(number: int, tuple_len: int) -> tuple:
    return tuple(int(binary) for binary in f"{number:b}".zfill(tuple_len))

def calculate_total_energy(sequence: tuple) -> int:
    return sum(calculate_autocorrelation(sequence, k_value)**2 for k_value in range(1, len(sequence)))

def calculate_autocorrelation(sequence: tuple, k_value: int) -> int:
    return sum(sequence[idx]*sequence[idx+k_value] for idx in range(len(sequence) - k_value))

def generate_sequences(seq_len: int) -> Tuple[Tuple]:
    return tuple(to_binary_tuple(number, seq_len) for number in range(0, 2 ** seq_len))

def solve_sequences(sequences: Tuple[Tuple]):
    return {sequence: calculate_total_energy(sequence) for sequence in sequences}
```

```python
sequences = solve_sequences(generate_sequences(8)) # 10 is safe xD
x = range(len(sequences))
y = sequences.values()
fig = plt.figure(figsize=(20,6))
plt.plot(x, y)
plt.show()
```
